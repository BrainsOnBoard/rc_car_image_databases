```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv('coordinates.csv')
df2 = pd.read_csv('coordinates1.csv')

plt.plot(df1['lat'], df1['lon'], df2['lat'], df2['lon'])
```

```python
import numpy as np
import utm

def gps2utm(df):
    easting, northing, *_ = utm.from_latlon(np.array(df['lat']), np.array(df['lon']))
    return (easting, northing)

(x1, y1) = gps2utm(df1)
(x2, y2) = gps2utm(df2)

plt.plot(x1, y1, x2, y2)
```
