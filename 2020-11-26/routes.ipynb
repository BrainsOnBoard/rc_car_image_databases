```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import utm

def gps2utm(df):
    easting, northing, *_ = utm.from_latlon(np.array(df['lat']), np.array(df['lon']))
    return (easting, northing)

datasets = ['dataset%i' % i for i in range(1, 4)]

_, ax = plt.subplots()
for dataset in datasets:
    df = pd.read_csv(dataset + '/coordinates.csv')
    ax.plot(*gps2utm(df))

ax.axis('equal')
ax.set_xlabel('Easting (m)')
ax.set_ylabel('Northing (m)')
ax.legend(datasets);
```
